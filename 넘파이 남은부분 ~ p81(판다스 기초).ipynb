선형대수 연산 - 행렬 내적, 전치행렬

```python
#행렬내적

a=np.array([[1,2,3],[4,5,6]])
b=np.array([[7,8],[9,10],[11,12]])

dot=np.dot(a,b)

print('내적 결과 :\n',dot)
```

```python
#전치행렬

a=np.array([[1,2],[3,4]])
transform=a.T

print('전치행렬 :\n',transform)
```
